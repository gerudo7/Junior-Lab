```python
#don't need to copy this; it's already in the other one
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
#Importing the T2 Data
raw2 = pd.read_csv('T2 Data.csv')
t2 = raw2.iloc[:,0].values
V2 = raw2.iloc[:,1].values
```

```python
plt.close('all')
plt.plot(t2,V2,"b.",label="Data")
plt.title("Voltage (V) vs. Tau (s) [90 then 180]")
plt.xlabel("Tau (s)")
plt.ylabel("Voltage (V)")
plt.legend()
#
#Define Function to Fit
#
def func(t2,V2max,T2):
    return V2max*np.exp(-2*t2/T2)
#
#Set Initial Guess of Fit Parameters and Curve Fit
#
popt,pcov=curve_fit(func,t2,V2,p0=(10,0.04))
print("V2max,T2",popt)
plt.plot(t2,func(t2,*popt),'r--',label='Fit: V2max = %3.3f volts,\
 T2 = %4.2f s' % tuple(popt))
plt.grid()
plt.legend()
```

```python
#Here's our output for standard deviation
perr = np.sqrt(np.diag(pcov))
print (perr)
```
