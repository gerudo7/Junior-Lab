```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#all measurements are in meters
b = 100e-6
l = 25e-2
lamb = 625e-9
h = 20e-3
step = 1e-6
```

```python
y = np.arange(start=-h/2,stop=h/2,step=step,dtype=None)
```

```python
d1 = np.zeros(len(y))
d2 = np.zeros(len(y))
A = np.zeros(len(d1))
I = np.zeros(len(A))
```

```python
i = 0
while i < len(y):
    d1[i] = math.sqrt((y[i]-b/2)**2+l**2)
    d2[i] = math.sqrt((y[i]+b/2)**2+l**2)
    i += 1
```

```python
i = 0
while i < len(y):
    A[i] = 2*math.cos(math.pi*(d1[i]-d2[i])/lamb)*math.cos(math.pi*(d1[i]+d2[i])/lamb)
    i += 1
```

```python
i = 0
while i < len(y):
    I[i] = A[i]**2
    i += 1
```

```python
plt.plot(y,I)
```
